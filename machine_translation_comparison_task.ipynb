```python
import json
import requests

from google import genai
from groq import Groq
import cohere as Cohere

from abc import ABC, abstractmethod
import pandas as pd
```

Для удобства выделим для каждой модели отдельный класс. Они все имеют единый интерфейс — метод `generate()`, который принимает промпт и возвращает ответ модели.

```python
class AIProvider(ABC):
  def __init__(self, name: str):
    self.name = name

  @abstractmethod
  def generate(self, prompt: str) -> str:
    pass
```

```python
class GeminiProvider(AIProvider):
  def __init__(self, api_key: str, model="gemini-2.5-flash"):
    super().__init__(name=f"Gemini ({model})")
    self.client = genai.Client(api_key=api_key)
    self.model = model

  def generate(self, prompt: str) -> str:
    response = self.client.models.generate_content(
      model=self.model,
      contents=prompt
    )
    return response.text
```

```python
class GroqProvider(AIProvider):
  def __init__(self, api_key: str, model="llama-3.3-70b-versatile"):
    super().__init__(name=f"Groq ({model})")
    self.client = Groq(api_key=api_key)
    self.model = model

  def generate(self, prompt: str) -> str:
    response = self.client.chat.completions.create(
      model=self.model,
      messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content
```

```python
class CohereProvider(AIProvider):
  def __init__(self, api_key: str, model="command-a-03-2025"):
    super().__init__(name=f"Cohere ({model})")
    self.client = Cohere.ClientV2(api_key)
    self.model = model

  def generate(self, prompt: str) -> str:
    response = self.client.chat(
      model=self.model,
      messages=[{"role": "user", "content": prompt}]
    )
    return response.message.content[0].text
```

Тексты будем подгружать из заранее определенного json файла, лежащего в репозитории проекта

```python
url = 'https://raw.githubusercontent.com/annakiryan/AI-WEEK/refs/heads/main/text.json'
texts_data = requests.get(url).json()
```

```python
gemini = GeminiProvider(GEMINI_API_KEY)
groq = GroqProvider(GROQ_API_KEY)
cohere = CohereProvider(COHERE_API_KEY)

models = [gemini, groq, cohere]
```

```python
translations = {}
```

Модель Gemini очень часто недоступна, поэтому при повторном запуске некоторые варианты текста могут быть не переведены для нее.

```python
MAX_RETRIES = 5

def safe_generate(model, prompt):
  for attempt in range(MAX_RETRIES):
    try:
      return model.generate(prompt)
    except Exception as e:
      print(f"{model.name} попытка {attempt+1}/{MAX_RETRIES} провалилась:", e)
  return None

for item in texts_data:
  prompt = prompt_base + item["text"]

  models_response = {
    model.name: safe_generate(model, prompt)
    for model in models
  }

  translations[item["id"]] = models_response
```

```python
# @title
from IPython.display import HTML, display

def show_columns(results_dict):
    """
    results_dict: { model_name: text }
    """
    html_cols = ""

    for model_name, text in results_dict.items():
        html_cols += f"""
        <div style="
            flex:1;
            padding:15px;
            border:1px solid #ccc;
            border-radius:10px;
            background:#fafafa;
            color:#000;
            white-space:pre-wrap;
        ">
            <h3 style="text-align:center; margin-top:0;color:#000;">{model_name}</h3>
            {text}
        </div>
        """

    html = f"""
    <div style="
        display:flex;
        gap:20px;
        width:100%;
        color:#000;
        margin:20px 0;
    ">
        {html_cols}
    </div>
    """

    display(HTML(html))
```

```python
for text_id, model_results in translations.items():
    print(f"Текст №{text_id}")
    show_columns(model_results)
```
